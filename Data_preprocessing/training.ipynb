```python
def convert_date(row):
    start = int(row.split('/')[0])
    return datetime(start, 1, 1)

def adjust_date(row):
    new_year = row['Fiscal Year'].year
    if row['Term Type'] == 'Winter term':
        new_month = 1 # january
        new_year += 1
    elif row['Term Type'] == 'Spring term':
        new_month = 5 # may
    elif row['Term Type'] == 'Fall term':
        new_month = 9 # september
    else:
        return row['Fiscal Year']
    return datetime(new_year, new_month, 1)
ds_copy['Fiscal Year'] = ds_copy['Fiscal Year'].apply(convert_date)
ds_copy['Fiscal Year'] = ds_copy.apply(adjust_date, axis=1)
ds_copy['Fiscal Year'] = pd.to_datetime(ds_copy['Fiscal Year']).dt.strftime('%Y-%m')
ds_copy
```

```python
ds_copy = ds_copy.drop('Term Type', axis=1)

ds_copy = ds_copy.rename(columns={'Coop Regular': 'Co-op'})
ds_copy['Co-op'] = ds_copy['Co-op'].replace({'Co-op': 1, 'Regular': 0})

ds_copy = ds_copy.rename(columns={'Work Term': 'Academic Term'})
ds_copy['Academic Term'] = ds_copy['Academic Term'].replace({'Academic Term': 1, 'Work Term': 0})

ds_copy = ds_copy.rename(columns={'Career': 'Undergraduate'})
ds_copy['Undergraduate'] = ds_copy['Undergraduate'].replace({'Undergraduate': 1, 'Graduate': 0})

ds_copy.sample(n=5)
```

```python
data_2024 = ds_copy[ds_copy['Fiscal Year'] == '2024-01'] # only use data from winter 2024 term
#data_2024.describe(include='all')
```

```python
weighted_counts = data_2024.groupby('Visa Status')['Student Headcounts'].sum()
total_students = weighted_counts.sum()
weighted_percentages = (weighted_counts / total_students) * 100
print(weighted_percentages)
labels = ['Canadian', 'Canadian Permanent Resident ','International']

fig, ax = plt.subplots()
ax.pie(weighted_percentages, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#999999'])
ax.set_title('Percentage of Undergraduate vs. Graduate Students')

plt.show()
```

```python
program_level_counts = data_2024.groupby('Program Level')['Student Headcounts'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=program_level_counts.index, y=program_level_counts.values, palette='rocket')
plt.title('Distribution of Students by Program Level')
plt.ylabel('Number of Students')
plt.xlabel('Program Level')
plt.xticks()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center',
                va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.show()
```

```python
coop_counts = data_2024.groupby('Co-op')['Student Headcounts'].sum()

labels = ['Regular', 'Co-op']

plt.pie(coop_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Proportion of Students in Co-op vs. Regular Programs')

plt.show()
```

```python
visa_status_counts = data_2024.groupby('Visa Status')['Student Headcounts'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=visa_status_counts.index, y=visa_status_counts.values, palette='rocket')
plt.title('Distribution of Students by Visa Status')
plt.ylabel('Number of Students')
plt.xlabel('Visa Status')
plt.xticks()

plt.show()
```

```python
pivot_table = ds_copy.pivot_table(index='Faculty (group)', columns='Fiscal Year', values='Student Headcounts', aggfunc='sum')

plt.figure(figsize=(20, 12))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='coolwarm', annot_kws={"size": 8})
plt.title('Heatmap of Program Enrollment by Term')
plt.ylabel('Program')
plt.xlabel('Term')

plt.show()
```

```python
grouped_data = data_2024.groupby(['Faculty (group)', 'Visa Status'])['Student Headcounts'].sum().unstack()
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='rocket')

plt.title('Visa Status Distribution by Faculty')
plt.ylabel('Number of Students')
plt.xlabel('Faculty')
plt.xticks(rotation=45)
plt.legend(title='Visa Status')

plt.show()
```

```python
ds_copy_group_object = ds_copy.groupby('customer_id')['trans_date'].max()
print(ds_copy_group_object)
```

```python
ds_newgroupeddata = pd.DataFrame(ds_copy_group_object)
```

```python
print(ds_newgroupeddata)
```

```python
ds_newgroupeddata['IsStoppedShopping'] = 0
```

```python
print(ds_newgroupeddata)
```

```python
import datetime as dt
datetobecompare = dt.datetime(2019,9,16)
ds_newgroupeddata.loc[ds_newgroupeddata['trans_date'] <= datetobecompare, 'IsStoppedShopping'] = 1
```

```python
ds_newgroupeddata
```

```python
ds_newgroupeddata["total_trns"] = ds_copy.groupby('customer_id').size()
ds_newgroupeddata["total_spends"] = ds_copy.groupby('customer_id')['tran_amount'].sum()
```

```python
ds_newgroupeddata
```

```python
ds_newgroupeddata = ds_newgroupeddata[ds_newgroupeddata['IsStoppedShopping']==1]
```

```python
ds_newgroupeddata
```

```python
ds_newgroupeddata.tail()
```

```python
ds_newgroupeddata['trans_date'].min()
```

```python
ds_newgroupeddata.describe()
```

```python
ds_newgroupeddata['scaled_transactions'] = (ds_newgroupeddata['total_trns'] - ds_newgroupeddata['total_trns'].min())\
                                           /(ds_newgroupeddata['total_trns'].max() - ds_newgroupeddata['total_trns'].min())
```

```python
ds_newgroupeddata.head()
```

```python
ds_newgroupeddata['spend_transactions'] = (ds_newgroupeddata['total_spends'] - ds_newgroupeddata['total_spends'].min())\
                                           /(ds_newgroupeddata['total_spends'].max() - ds_newgroupeddata['total_spends'].min())
```

```python
ds_newgroupeddata.head()
```

```python
ds_newgroupeddata['score'] = ((0.5 * ds_newgroupeddata['scaled_transactions']) + (0.5 * ds_newgroupeddata['spend_transactions'])) * 100
```

```python
ds_newgroupeddata.head()
```

```python
cutoff2 = dt.datetime(2019,1,1)
```

```python
final_cleaned_data = ds_newgroupeddata[ds_newgroupeddata['trans_date'] >= cutoff2 ]
final_cleaned_data = final_cleaned_data.sort_values(by = ['trans_date'])
final_cleaned_data.head()
```

```python
final_cleaned_data.drop(columns = ['IsStoppedShopping','trans_date','total_trns','total_spends'], inplace = True)
```

```python
final_cleaned_data.head()
```

```python
final_cleaned_data.sort_values(['score'], ascending = False, inplace = True)
```
